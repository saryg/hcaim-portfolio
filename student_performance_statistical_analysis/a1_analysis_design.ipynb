# Analysis Design

This analysis design document describes the study of student performance in Mathematics and Portuguese language in secondary education in Portugal, during the 2005-2006 school year.

## Subject domain of interest and research questions

Student performance and the underlying issues causing poor performance are very important to a wide number of people. This information is valuable to students, their parents, teachers, schools, and governmental education departments. A variety of factors are thought to affect student's grades from socio-economic status to an individual's lifestyle and personal health.

The work described here was carried out to investigate what factors affect student success in secondary school. The research question is:

**What factors, if any, affect student success in secondary school grades?**

## Conceptual Framework

The cases of interest are **student profiles**.

**Grades** are a common way of measuring student performance. Over a period of time grades can be calculated on multiple occasions giving more than one snapshot of a student's capability at that time.  

Performance is expected to depend on a wide-range of factors therefore a number of representative variables have been chosen to be examined because they may contribute to how well the student performs.  

The variables upon which the performance is expected to depend are: 
- the amount of **study time** a student puts into understanding the class material,
- whether the student has **failed a class previously**, 
- the student's **reason** for attending this school,
- the student's **travel time** to the school,
-  if the student has attended extra **paid** classes,
- the **sex** of the student,
- the quality of their **family relationships**,
- whether their **parents are together**,
- the amount of **free time** the student has,
- how often the student drinks **alcohol during the week**,
- how often the student drinks **alcohol at the weekend**,
- how often the student **goes out with friends**,
- whether the student lives in a rural or urban **location**,
- whether their primary **guardian** is their mother or father, or someone else,
- how highly educated are their **mother** and **father**,
- what type of **job** their parents have,
- whether the student attended **nursery**,
- whether the student has **internet** at home,
- how **supportive** their parents are,
- how good the student's **health** is,
- how many **absences** the student has from class.

It is expected that a more stable and supportive home life would have a positive effect on a student's success, as would being in good health. Having well educated parents is likely an advantage for a student as their parents may be able to provide additional assistance at home.

It would be expected that students who spend more time studying will have better grades, and taking extra classes or internet resources may allow poorer students to catch up and improve their grades. However, if the student has failed classes previously they might not be strong in that subject or there may be other underlying issues. Similarly. students who have chosen to study at this particular school may be more motivated to perform well than those who attend because it is required.

A lack of time might have a negative effect on the student's grades. For example, missing too much class time, or being too tired due to a lack of free time, excessive travel time to the school, or free time spent socialising with friends. The student's location and distance from a town may be linked to these factors.

## Theory

Students' grades vary based on factors other than ability alone.  


## Hypotheses

>Students who have more support at home are associated with better grades.

>Students who drink alcohol more frequently during the week are associated with worse grades.

> Better previous grades will be associated with better final grades. Students who have performed well in the past will perform well in the final grade.

> Students who have more free time are associated with better final grades, but this will vary by sex.

## Data
The data were collected during the 2005-2006 school year from two public schools in the Alentejo region of Portugal. The student data obtained was focused on Maths and Portuguese language classes.  Grades and absences were sourced from school grade reports, while all other data was obtained through a 37-question questionnaire given to the students. 
The questionnaire was composed of closed questions relating to the student's demographics, their lifestyle, and some school based questions.
Data from the school reports included 3 periodic grades (of which one was the final grade) and the number of absences from the class.

## Design Evaluation

**Validity**:
- **External**: Threatened by the fact that data was sourced from only 2 schools, in the same region of Portugal. The study may not generalise well unless these schools are representative of the country as a whole. More data from more schools may be needed to get a representative sample.
- **internal** - Threatened by the voluntary nature of questionnaire responses. Some responses to the questionnaire were lacking data and were therefore discarded. It's not known if the discarded data is from students who may have tended to be high or low performers.
- **construct** - Threatened by the assumption that a grade is only dependent on a student's effort. Good teachers may be able to teach weaker students more effectively. The exams that the grades are based on may not be representative of the subject as a whole and may only be concerned with a single topic. 
- **content** - Threatened by breadth of the possible factors that may influence a student's school performance. It is likely that very many factors have an affect other than the factors identified. e.g. diet, hours of sleep, etc.


**Reliability**:

- The data was in part collected via questionnaires. Although efforts were made to include only closed questions, the voluntary nature of a questionnaire means that students may not answer all questions, or may answer in a subjective manner, meaning the same responses might not be obtained if the questionnaire was repeated. This particularly applies to subjective questions such as the frequency of alcohol on weekends being scored from 1 to 5.
- The data was collected from only 2 schools in the same region, therefore it is not obvious if the data will apply more generally.
