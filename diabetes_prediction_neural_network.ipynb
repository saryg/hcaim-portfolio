```python
import numpy as np

# Random seed already set to 1

# Shuffle the dataset
shuffled_indices = np.random.permutation(len(df))
data_shuffled = df.iloc[shuffled_indices]

X = data_shuffled.iloc[:, :-1]  # Features: all columns except the last one
y = data_shuffled.iloc[:, -1]   # Target variable: the last column

#Normalise X values to be between 0 and 1
scaler = sklearn.preprocessing.Normalizer().fit(X)
normalizedX = scaler.transform(X)
normalizedX

# Calculate split indices
train_size = int(0.79 * len(X))
validation_size = int(0.2 * len(X))
test_size = int(0.01 * len(X)) #len(df) - train_size - validation_size

# Split the data
train_data = X.iloc[:train_size]
validation_data = X.iloc[train_size:train_size + validation_size]
test_data = X.iloc[train_size + validation_size:]

# Separate features and target for each set
X_train = train_data
y_train = y.iloc[:train_size]

X_val = validation_data
y_val = y.iloc[train_size:train_size + validation_size]
X_test = test_data
y_test = y.iloc[train_size + validation_size:] #test_data['Diabetes_binary']

# Check the size of each set
X_train.shape, X_val.shape, X_test.shape
```

# 2. Model Exploration (8 networks)

- Capacity - number of nodes per layer
- Depth - number of layers


Rationale:
- test small initial and hidden layer sizes, if they underfit increase the size of the layer capacity. Avoid making the layers so big that the model starts to overfit and memorise the data.

- test increasing depths starting at 1 hidden layer.

- simple baseline - start with 50 nodes in the initial layer and 50 in the hidden layer
- test depth by increasing the number of layers up to 4 and inspecting the plots
- test width by increasing the capacity of the initial and hidden layers and inspecting the plots.

These were tested with batch size = 500

```python
# set up architectures to test

model_architectures = [# initial layer, hidden layers
          (50, 50),
          (50, 50, 25),
          (50, 50, 50, 25),
          (100,),
          (100, 100),
          (100, 100, 100),
          (500, 500, 100),
          (1000, 100, 50),
          (1000, 100, 100),
          (1000, 100, 100, 50)
          ]

def create_model(model_architecture, dropout = False):

  model = Sequential()
  dropout_pc = 0.2 # 20% dropout rate

  # initial layer
  if dropout==True:
    model.add(Dropout(dropout_pc, input_shape =(X_train.shape[1],)))
  model.add(Dense(model_architecture[0], input_dim=X_train.shape[1], activation='relu'))
  model_desc = f"Architecture\ninitial layer:{model_architecture[0]}"
  plot_text = f"({model_architecture[0]}"

  # hidden layers
  for i, dimension in enumerate(model_architecture[1:]):
    if dropout==True:
      model.add(Dropout(dropout_pc, input_shape =(X_train.shape[1],)))

    model.add(Dense(dimension, activation='relu'))
    model_desc = model_desc + f"\nHidden layer {i+1}: {dimension}"
    plot_text = plot_text+ f", {dimension}"

  # output layer
  if dropout==True:
    model.add(Dropout(dropout_pc, input_shape =(X_train.shape[1],)))
  model.add(Dense(1, activation='sigmoid'))
  model_desc = model_desc + f"\nOutput layer: 1 \n\n"
  plot_text = plot_text+ f", 1)"

  return model, model_desc, plot_text
```

```python
# test each architecture setup
models_results_history = []
for i, model_architecture in enumerate(model_architectures):

  # fix tensorflow random seed for reproducibility
  #seed=1 set previously
  np.random.seed(seed)
  tf.random.set_seed(seed) # for TF this needs to be run per cell.


  model, model_desc, plot_text = create_model(model_architecture, dropout=False)
  print("Model ", i, model_desc)

  # Compile model
  model.compile(loss='binary_crossentropy', optimizer="adam", metrics=['acc'])

  n_epochs = 50

  # Fit the model, but this time store the history (that is accuracy and loss for each Epoch)
  results = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=n_epochs, batch_size=500, verbose=verbose_switch, shuffle=True)

  models_results_history.append((plot_text, results))

  fig, axs = plt.subplots(1, 2, figsize = (12, 6))

  for ax in axs:
    ax.set_xlim([0, n_epochs])
    ax.set_ylim([0.45, 0.8])


  # summarize history for loss
  axs[0].plot(results.history['loss'])
  axs[0].plot(results.history['val_loss'])
  axs[0].set_title(f'model loss - {plot_text}')
  axs[0].set_ylabel('loss')
  axs[0].set_xlabel('epoch')
  axs[0].legend(['train loss', 'val loss'], loc='upper left')


  # summarize history for accuracy
  axs[1].plot(results.history['acc'])
  axs[1].plot(results.history['val_acc'])
  axs[1].set_title(f'model accuracy - {plot_text}')
  axs[1].set_ylabel('accuracy')
  axs[1].set_xlabel('epoch')
  axs[1].legend(['train loss', 'val loss'], loc='upper left')


```

```python
# to collect the model acc and val_acc history for comparison

def make_model_history_df(models_results_hist):
  rows = []
  for model_variant, model_results in models_results_hist:
    #start_epoch = start #epochs to exclude from averaging accuracy
    #end_epoch = n_epochs

   # mean_acc = sum(model_results.history['acc'][start_epoch:end_epoch]) / len(model_results.history['acc'][start_epoch:end_epoch])
    #mean_val_acc = sum(model_results.history['val_acc'][start_epoch:end_epoch]) / len(model_results.history['val_acc'][start_epoch:end_epoch])
    row = {"model_variant": model_variant,
          "max_val_acc": max(model_results.history["val_acc"]),
          "max_acc": max(model_results.history["acc"]),
          #"mean_acc": mean_acc,
          #"mean_val_acc": mean_val_acc,
          "min_loss": min(model_results.history["loss"]),
          "min_val_loss": min(model_results.history["val_loss"])
          }

    rows.append(row)


  results_df = pd.DataFrame(rows)
  return results_df


results_df = make_model_history_df(models_results_history )
results_df
```

Overall the different architectures performed quite similarly. There did not appear to be any major advantages for any of the model architectures tested.

Increasing the capacity from 50 to 100 to 500 to 10000 does not seem to have improved or weakened the model's performance.

Increasing the depth also does not seem to have improved the model's performance.

From the performance plots above it appears that increasing the capacity from 50 to 100 increases noise in the plots, and further increasing the capacity to 500 and 1000 makes the noise even more apparent.

Increasing the depth seems to amplify any noise introduced by the increasing capacities.

All models converge quite quickly and have a similar performance. The smaller capacity models have smoother curves and are more stable on the validation sets in comparison with the larger and deeper models. None of the models seem to over or under fit. although the larger models become more erratic as they grow.

Comparing the max accuracy on the train and validation sets the combination of 100,1 looks like a reasonable architecture to use because it's not overly complex or deep without any benefits and it has a very similar performance to the more complex structures, particularly on the validation set.

Following on from the model structure exploration, the best model architecture will be used when tuning the hyperparameters.

#### Batch sizes
A range of batch sizes will be tested ranging from very small (1) to very large (5000). As the batch sizes are increased the number of epochs will also be increased to allow the model time to reach convergence as the model waits longer before updating the weights.

It is expected that smaller batch sizes will cause more noise than larger sizes.

#### Optimizers
The optimizers to be tested are RMSprop, Adagrad, and Adam. These adapt the learning rate depending on performance and help to speed up training and convergence by dynamically changing how much a weight will be changed by at each step.

#### Regularisation
The regularisation techniques to be applied are dropout and earlystopping. Dropout is used to randomly remove nodes from the network. Early stopping prevents the model from overtraining due to running for too long.

```python
def subMetrics(targetGroup, act, pred):
  print("Metrics for: ", targetGroup)
  correct = 0
  # parse both lists to numpy arrays
  act = np.array(act)
  pred = np.array(pred)
  # There may be an easier way to do this, however, I believe it is a manual process!
  for index, item in enumerate(act):
    if item == pred[index]:
      correct += 1
  # print accuracy, and the confusion matrix
  accuracy = round((correct / len(act)) * 100,2)
  print("Accuracy:", accuracy, "%")
  print(pd.crosstab( np.array(act),  np.array(pred), rownames=['Actual'], colnames=['Predicted'], margins=True))
```

```python
X_test['y_pred'] = y_pred # add the y predicted class to the x_test df to do comparisons
X_test['y_test'] = y_test # add the actual y class
```

```python

# filter the DataFrame for females and males
females = X_test[X_test.iloc[:, 17] == 0]  # Assuming the gender column is at index 17
males = X_test[X_test.iloc[:, 17] == 1]

# predictions and actual classes for females and males
female_pred = females['y_pred'].tolist()
female_actual_class = females['y_test'].tolist()

male_pred = males['y_pred'].tolist()
male_actual_class = males['y_test'].tolist()

# Print the representation of each target group
print("Female instances in the test set:", len(female_pred))
print("Male instances in the test set:  ", len(male_pred))


print("\n"*2)
# Confusion Matrix and classification report
subMetrics("Female", female_actual_class, female_pred)
print("\n")
subMetrics("Male", male_actual_class, male_pred)

print("\n")
print("\n")
print("Classification Report for Diabetes in Female group:")
cr = classification_report(female_actual_class, female_pred)
print(cr)

print("Classification Report for Diabetes in Male group:")
cr = classification_report(male_actual_class, male_pred)
print(cr)
```

Female instances were equally likely to be correctly identified as either diabetic or non-diabetic. Male instances had better precision, but worse recall if the were not diabetic than if they were.

Female predictions were much more balanced than male predictions even though the f1 scores are almost identical.
